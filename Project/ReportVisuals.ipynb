## Visuals and Data for Report ##

### Code for exploring the results to be used for the final report ###

**Importing libraries**

```python
import pandas as pd; pd.set_option('display.max_column', 100)
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

**Best Logistic Regression Model**

```python
lr_df = pd.read_csv('lr_results.csv')
lr_df_avgs = lr_df.groupby(['cut_off']).mean().reset_index(drop=False)
lr_df_avgs['model_type']='LogisticRegression'
lr_df_avgs['model_name']=lr_df_avgs['model_type'].astype(str)
lr_df_avgs
```

```python
lr_best = lr_df_avgs.sort_values('overall_score',ascending =False).reset_index(drop = True).head(1)
lr_best
```

**Best Support Vector Machine Model**

```python
svm_df = pd.read_csv('svm_results.csv')
svm_df_avgs = svm_df.groupby(['kernal','cut_off']).mean().reset_index(drop=False)
svm_df_avgs['model_type']='SupportVectorMachine'
svm_df_avgs['model_name']=svm_df_avgs['model_type'].astype(str)+'_'+svm_df_avgs['kernal'].astype(str)
svm_df_avgs
```

```python
svm_best = svm_df_avgs.sort_values('overall_score',ascending =False).reset_index(drop = True).head(1)
svm_best
```

**Best Decision Tree Model**

```python
dt_df = pd.read_csv('dt_results.csv')
dt_df_avgs = dt_df.groupby(['depth','cut_off']).mean().reset_index(drop=False)
dt_df_avgs['model_type']='DecisionTree'
dt_df_avgs['model_name']=dt_df_avgs['model_type'].astype(str)+'_'+dt_df_avgs['depth'].astype(str)
dt_df_avgs
```

```python
dt_best = dt_df_avgs.sort_values('overall_score',ascending =False).reset_index(drop = True).head(1)
dt_best
```

**Best Random Forest Model**

```python
rf_df = pd.read_csv('rf_results.csv')
rf_df_avgs = rf_df.groupby(['depth','n_tree','cut_off']).mean().reset_index(drop=False)
rf_df_avgs['model_type']='RandomForest'
rf_df_avgs['model_name']=rf_df_avgs['model_type'].astype(str)+'_'+rf_df_avgs['depth'].astype(str)+'_'+rf_df_avgs['n_tree'].astype(str)
rf_df_avgs
```

```python
rf_best = rf_df_avgs.sort_values('overall_score',ascending =False).reset_index(drop = True).head(1)
rf_best
```

**Best AdaBoost Model**

```python
ada_df = pd.read_csv('ada_results.csv')
ada_df_avgs = ada_df.groupby(['depth','n_tree','learning_rate','cut_off']).mean().reset_index(drop=False)
ada_df_avgs['model_type']='AdaBoost'
ada_df_avgs['model_name']=ada_df_avgs['model_type'].astype(str)+'_'+ada_df_avgs['depth'].astype(str)+'_'+ada_df_avgs['n_tree'].astype(str)+'_'+ada_df_avgs['learning_rate'].astype(str)
ada_df_avgs
```

```python
ada_best = ada_df_avgs.sort_values('overall_score',ascending =False).reset_index(drop = True).head(1)
ada_best
```

**Best Gradient Boosting Model**

```python
gb_df = pd.read_csv('gb_results.csv')
gb_df_avgs = gb_df.groupby(['depth','n_tree','learning_rate','cut_off']).mean().reset_index(drop=False)
gb_df_avgs['model_type']='GradientBoosting'
gb_df_avgs['model_name']=gb_df_avgs['model_type'].astype(str)+'_'+gb_df_avgs['depth'].astype(str)+'_'+gb_df_avgs['n_tree'].astype(str)+'_'+gb_df_avgs['learning_rate'].astype(str)
gb_df_avgs
```

```python
gb_best = gb_df_avgs.sort_values('overall_score',ascending =False).reset_index(drop = True).head(1)
gb_best
```

**Best Model for each type**

```python
lr = lr_best[['model_type','cut_off','accuracy','recall','f1','overall_score','model_name']]
svm = svm_best[['model_type','cut_off','accuracy','recall','f1','overall_score','model_name']]
dt = dt_best[['model_type','cut_off','accuracy','recall','f1','overall_score','model_name']]
rf = rf_best[['model_type','cut_off','accuracy','recall','f1','overall_score','model_name']]
ada = ada_best[['model_type','cut_off','accuracy','recall','f1','overall_score','model_name']]
gb = gb_best[['model_type','cut_off','accuracy','recall','f1','overall_score','model_name']]

best_models = pd.concat([lr,svm,dt,rf,ada,gb])
best_models.sort_values('overall_score',ascending =False).reset_index(drop = True)
```

Best model is Logistic Regression with a 0.35 cut off

**Evaluating Cutoffs**

Evaluating the cutoff value for each of the best models

```python
lr_df_avgs['best_model'] = np.where(lr_df_avgs['model_name'] == 'LogisticRegression',1,0)
svm_df_avgs['best_model'] = np.where(svm_df_avgs['model_name'] == 'SupportVectorMachine_rbf',1,0)
dt_df_avgs['best_model'] = np.where(dt_df_avgs['model_name'] == 'DecisionTree_3',1,0)
rf_df_avgs['best_model'] = np.where(rf_df_avgs['model_name'] == 'RandomForest_7_300',1,0)
ada_df_avgs['best_model'] = np.where(ada_df_avgs['model_name'] == 'AdaBoost_3_1000_0.001',1,0)
gb_df_avgs['best_model'] = np.where(gb_df_avgs['model_name'] == 'GradientBoosting_3_1000_0.001',1,0)
```

```python
lr_cut = lr_df_avgs[lr_df_avgs['best_model']==1]
svm_cut = svm_df_avgs[svm_df_avgs['best_model']==1]
dt_cut = dt_df_avgs[dt_df_avgs['best_model']==1]
rf_cut = rf_df_avgs[rf_df_avgs['best_model']==1]
ada_cut = ada_df_avgs[ada_df_avgs['best_model']==1]
gb_cut = gb_df_avgs[gb_df_avgs['best_model']==1]
```

```python
plt.plot(lr_cut['cut_off'],lr_cut['accuracy'])
plt.plot(svm_cut['cut_off'],svm_cut['accuracy'])
plt.plot(dt_cut['cut_off'],dt_cut['accuracy'])
plt.plot(rf_cut['cut_off'],rf_cut['accuracy'])
plt.plot(ada_cut['cut_off'],ada_cut['accuracy'])
plt.plot(gb_cut['cut_off'],gb_cut['accuracy'])
plt.xlabel('Cut Off')
plt.xticks(np.linspace(0.05,.5,num=10))
plt.ylabel('Accuracy')
plt.grid(linestyle='--',axis='y')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, labels = ['LogisticRegression','SupportVectorMachine_rbf','DecisionTree_3','RandomForest_7_300',
                                        'AdaBoost_3_1000_0.001','GradientBoosting_3_1000_0.001'])
plt.show();
```

```python
plt.plot(lr_cut['cut_off'],lr_cut['recall'])
plt.plot(svm_cut['cut_off'],svm_cut['recall'])
plt.plot(dt_cut['cut_off'],dt_cut['recall'])
plt.plot(rf_cut['cut_off'],rf_cut['recall'])
plt.plot(ada_cut['cut_off'],ada_cut['recall'])
plt.plot(gb_cut['cut_off'],gb_cut['recall'])
plt.xlabel('Cut Off')
plt.xticks(np.linspace(0.05,.5,num=10))
plt.ylabel('Recall')
plt.grid(linestyle='--',axis='y')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, labels = ['LogisticRegression','SupportVectorMachine_rbf','DecisionTree_3','RandomForest_7_300',
                                        'AdaBoost_3_1000_0.001','GradientBoosting_3_1000_0.001'])
plt.show();
```

```python
plt.plot(lr_cut['cut_off'],lr_cut['f1'])
plt.plot(svm_cut['cut_off'],svm_cut['f1'])
plt.plot(dt_cut['cut_off'],dt_cut['f1'])
plt.plot(rf_cut['cut_off'],rf_cut['f1'])
plt.plot(ada_cut['cut_off'],ada_cut['f1'])
plt.plot(gb_cut['cut_off'],gb_cut['f1'])
plt.xlabel('Cut Off')
plt.xticks(np.linspace(0.05,.5,num=10))
plt.ylabel('F1 Score')
plt.grid(linestyle='--',axis='y')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, labels = ['LogisticRegression','SupportVectorMachine_rbf','DecisionTree_3','RandomForest_7_300',
                                        'AdaBoost_3_1000_0.001','GradientBoosting_3_1000_0.001'])
plt.show();
```
